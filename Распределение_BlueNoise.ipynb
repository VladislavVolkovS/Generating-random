```python
import numpy as np
import pandas as pd
```

```python
csv = pd.read_csv("blue_noise_res.csv", delimiter=',', header=None)
```

```python
a = np.array(csv)
```

```python
import matplotlib.pyplot as plt

plt.title("blue_noise_r=0.02")
plt.xlim(0, 1)
plt.ylim(0, 1)
plt.grid(True, color = "grey", linewidth = "1", alpha = 0.1)
for i in range(a.shape[0]):
  plt.plot(a[i][0], a[i][1], 'ro', markersize=1)
plt.savefig("distribution_blue_noise_" + str(a.shape[0]) + "_r=0.02.png")
```
